````python
#Загрузка данных
X=pd.read_csv("https://drive.google.com/uc?id=1i429c9_NemRt6Kga_ApRGpext8vERO0y")

#Кодируем данные в понятный моделе формат
encoder = OneHotEncoder(sparse_output=False)
X_dum = encoder.fit_transform(X)

#применение модели к данным
y_predicted = СBoost.predict(X_dum)
y = pd.DataFrame(list(y_predicted))
y.columns=['account_status']
result = pd.merge(X, y, left_index=True, right_index=True, how='outer')
result
````

##*Применение модели при ручном вводе данных:*

````python
#predict =[2,'M','Y','Y','Отсутствуют','10_quantile','Working','Highereducation','Civilmarriage','Rentedapartment','Есть','Есть','Есть','Есть','Неизвестно','Два','10-15лет','30-40']
exemples = pd.read_csv("https://drive.google.com/uc?id=1gVLR6m0hicsnFB4sae534ADGKZ7fhPvC")    #Загружаем шаблон со списоком всех категорий, для корректной обработки моделью.

#Вводим значения по циклу
predict=[]
for i in exemples.columns:
    print(f'Введите {i}')
    item = input()
    predict.append(item)

#Дописываем введенные значения в датафрейм
exemples.loc[len(exemples)] = predict
#exemples

#Изменяем тип столбцов на object
for column in exemples.columns:
    exemples[column] = exemples[column].astype('str')

#Кодируем данные в понятный моделе формат
encoder = OneHotEncoder(sparse_output=False)
X_dum_1 = encoder.fit_transform(exemples)

#применение модели к данным
y_predicted = СBoost.predict(X_dum_1)
y = pd.DataFrame(list(y_predicted))
y.columns=['account_status']
result = pd.merge(exemples, y, left_index=True, right_index=True, how='outer')

#вывод результата
print('\nВ результате анализа модели прогнозируется следующее состояние по просрочкам (0: "Вероятно просрочек не будет", 1: "вероятна просрочка")\n', result['account_status'].tail(1))
````
